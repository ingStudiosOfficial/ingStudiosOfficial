```python
print("Prime Number Checker")
start_num = input(prompt = "Start Number: ")
end_num = input(prompt = "End Number: ")
prime = 0
not_prime = 0
for i in range(int(start_num),int(end_num)+1):
  print("Current Number: " + str(i))
  current_num = i
  if current_num == 1:
    print("Not Prime")
    print("--NEXT--")
    not_prime += 1
    continue
  elif current_num == 2:
    print("Prime")
    print("--NEXT--")
    prime += 1
    continue
  else:
    for l in range(2,current_num):
      if current_num%l == 0:
        print("Not Prime")
        not_prime += 1
        break
    else:
      print("Prime")
      prime += 1
    if current_num != int(end_num):
      print("--NEXT--")
    else:
      print("--END--")
percentage_prime = (prime/(prime+not_prime))*100
print("Prime Numbers: " + str(prime))
print("Not Prime Numbers: " + str(not_prime))
print(str(percentage_prime)+"% of the numbers calculated are prime!")
```
